```python
##Function to take value from historical position and find second to last item
def calc_prev_pos(hist_pos):
    values = hist_pos.split(', ')
        
    if len(values) == 1:
        prev_pos = values[0]
        
    else:
        prev_pos = values[-2]
    
    prev_pos = int(prev_pos)
    
    return prev_pos
```

```python
##apply function to hist_pos column
df['prev_pos'] = df['hist_pos'].apply(calc_prev_pos)
df[['hist_pos','prev_pos']][:10]
```

```python
##function to find popularity
def find_pop(track):
    track = sp.track(track)
    pop = track['popularity']
    
    return pop
    
##function to find uri    
def find_track_uri(track):
    track = sp.track(track)
    uri = track['uri']
    
    return uri
    

##function to find release date
def find_release_date(track):
    track = sp.track(track)
    
    ##release date from album
    album = track['album']
    release_date = album['release_date']
        
    return release_date
```

```python
##for each track find label/distributor
def find_label(track_id):
    track = sp.track(track_id)
    album = track['album']
    album_id = album['id']

    album = sp.album(album_id)
    label = album['label']
    
    return label
```

```python
##function to find artists popularity score
def find_artist_popularity(track_id):
    track = sp.track(track_id)
    
    ##get list of artists on track
    artists = track['artists']
    
    ##empty list for artist ids
    artist_ids = []
    
    ##find artist id for each artist in list
    for artist in artists:
        a_id = artist['id']
        
        ##add id to list
        artist_ids.append(a_id)
    
    ##empty list of popularity
    artist_popularities = []
    
    ##look up each artist by id and find popularity
    for a_id in artist_ids:
        artist = sp.artist(a_id)
        popularity = artist['popularity']
        
        ##add to list of popularity scores
        artist_popularities.append(popularity)
    
    return artist_popularities
```
